## 数据说明

- 文件：`data/orders.csv`（150 条德邦向示例运单）
- 关键字段：`revenue`（营收）、`cost`（成本）、`region`（区域）

**动手方式：** 从上到下，每个代码单元格点左侧 ▶ 运行（或 `Shift + Enter`）

````python
# 第 1 步：导入 Pandas（数据分析的核心工具）
import pandas as pd

print("Pandas 加载成功 ✅")
````

````python
# 第 2 步：读取运单数据
# 提示：read_csv 就像 Excel 的「打开文件」
df = pd.read_csv("data/orders.csv")

print(f"共加载 {len(df)} 条运单")
df.head(3)  # 看前 3 行，确认数据长什么样
````

````python
# 第 3 步：看看有哪些列（字段）
print("数据包含以下字段：")
print(df.columns.tolist())
````

---
## 第 4 步：计算 3 个核心 KPI ⭐

| KPI | 公式 | 经营含义 |
|-----|------|---------|
| **总营收** | 所有运单 revenue 之和 | 网点做了多少流水 |
| **总单量** | 运单条数 | 业务活跃度 |
| **毛利率** | (总营收 - 总成本) / 总营收 × 100% | 赚的是真钱还是流水 |

````python
# 先算每一单的毛利（营收 - 成本）
df["毛利"] = df["revenue"] - df["cost"]

# KPI 1：总营收
total_revenue = df["revenue"].sum()

# KPI 2：总单量
total_orders = len(df)

# KPI 3：毛利率
total_profit = df["毛利"].sum()
margin_rate = total_profit / total_revenue * 100

print("=" * 40)
print("  德邦网点 KPI 总览 · 海豚生")
print("=" * 40)
print(f"  总营收:   {total_revenue:,.2f} 元")
print(f"  总单量:   {total_orders} 单")
print(f"  毛利率:   {margin_rate:.2f} %")
print("=" * 40)
````
